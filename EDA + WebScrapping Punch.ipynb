# Cocktail Scrapper function

```python
def cocktailScraper():
    for i in url:
        response = get(i)
        soup = BeautifulSoup(response.content, 'html.parser')
        try:
            name = soup.find("h1", {"class": "entry-title text-center"})
            name_data = [name.text.strip('\n')]
            print(name)
        except:
            pass
        try:
            #ingredients = soup.find("li", {"itemprop": "ingredients"})
            ingredients = soup.find("ul", {"class": "direction-list"})
                #print(ingredients)
        except:
            pass
        try:
            garnish = soup.find("p",{'class':'garn-glass'})
                #print(garnish)
        except:
            pass
        try:
            directions = soup.find('ol')
                #print(directions)
        except:
            pass

        punchInfo = [name,ingredients,garnish,directions]
        row = []
        for info in punchInfo:
            row.append(info.get_text().strip())
        print(row)

        c.writerow( row )
        print('Success')
```
